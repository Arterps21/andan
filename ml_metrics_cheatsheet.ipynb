# Шпаргалка по метрикам машинного обучения

Подробный конспект по основным семействам метрик: регрессия, классификация, ранжирование, кластеризация, сегментация, прогнозирование и калибровка.

## 1. Как выбирать метрику

Метрика зависит от типа задачи, формы целевой переменной, важности разных ошибок и бизнес-смысла оценки.

Обычно различают:
- регрессию;
- классификацию;
- ранжирование и поиск;
- кластеризацию;
- сегментацию изображений;
- прогнозирование временных рядов;
- вероятностный прогноз;
- калибровку вероятностей.

## 2. Регрессия

Здесь целевая переменная — число: цена, время, просмотры, спрос, температура и т. п.

Обозначения:
- $y_i$ — истинное значение;
- $\hat y_i$ — предсказание;
- $n$ — число объектов;
- ошибка $e_i = y_i - \hat y_i$.

### 2.1 MAE — Mean Absolute Error

$$
MAE = \frac{1}{n}\sum_{i=1}^{n}|y_i - \hat y_i|
$$

Средняя абсолютная ошибка. Интерпретируемая и более устойчивая к выбросам, чем MSE.

### 2.2 MSE — Mean Squared Error

$$
MSE = \frac{1}{n}\sum_{i=1}^{n}(y_i - \hat y_i)^2
$$

Средняя квадратичная ошибка. Сильнее штрафует большие промахи.

### 2.3 RMSE — Root Mean Squared Error

$$
RMSE = \sqrt{\frac{1}{n}\sum_{i=1}^{n}(y_i - \hat y_i)^2}
$$

Корень из MSE. Возвращает ошибку в тех же единицах, что и целевая переменная.

### 2.4 Median Absolute Error

$$
MedAE = median(|y_i - \hat y_i|)
$$

Медианная абсолютная ошибка. Устойчива к выбросам.

### 2.5 R² — коэффициент детерминации

$$
R^2 = 1 - \frac{\sum_i (y_i - \hat y_i)^2}{\sum_i (y_i - \bar y)^2}
$$

Показывает, насколько модель лучше предсказания среднего значения.

### 2.6 Adjusted R²

$$
AdjR^2 = 1 - (1-R^2)\frac{n-1}{n-p-1}
$$

Штрафует за добавление лишних признаков, где $p$ — число признаков.

### 2.7 MAPE, sMAPE, WAPE

$$
MAPE = \frac{100\%}{n}\sum_{i=1}^{n}\left|\frac{y_i - \hat y_i}{y_i}\right|
$$

$$
sMAPE = \frac{100\%}{n}\sum_{i=1}^{n}\frac{2|y_i - \hat y_i|}{|y_i| + |\hat y_i|}
$$

$$
WAPE = \frac{\sum_i |y_i - \hat y_i|}{\sum_i |y_i|}
$$

Эти метрики удобны, когда важна относительная ошибка.

### 2.8 MSLE и RMSLE

$$
MSLE = \frac{1}{n}\sum_{i=1}^{n}(\log(1+y_i)-\log(1+\hat y_i))^2
$$

$$
RMSLE = \sqrt{MSLE}
$$

Полезны, когда важнее ошибка в лог-шкале.

### 2.9 Huber loss

$$
L_\delta(e)=
\begin{cases}
\frac{1}{2}e^2, & |e|\le \delta \\
\delta(|e|-\frac{1}{2}\delta), & |e|>\delta
\end{cases}
$$

Компромисс между MSE и MAE.

### 2.10 Quantile loss

$$
L_q(e)=
\begin{cases}
q\cdot e, & e\ge 0 \\
(q-1)\cdot e, & e<0
\end{cases}
$$

Используется для квантильного прогноза, интервалов и медианы.

### 2.11 Метрики для счётчиков

- Poisson deviance
- Gamma deviance
- Tweedie deviance

Они особенно полезны для неотрицательных величин и счётных данных.

## 3. Классификация

Здесь предсказывается класс: 0/1, несколько классов или несколько меток одновременно.

Для бинарной классификации используют матрицу ошибок:
- TP — true positive;
- FP — false positive;
- TN — true negative;
- FN — false negative.

### 3.1 Accuracy

$$
Accuracy = \frac{TP+TN}{TP+TN+FP+FN}
$$

Доля верных ответов. Плохо работает при дисбалансе классов.

### 3.2 Precision

$$
Precision = \frac{TP}{TP+FP}
$$

Из всех предсказанных положительных сколько действительно положительные.

### 3.3 Recall / Sensitivity / TPR

$$
Recall = \frac{TP}{TP+FN}
$$

Из всех реальных положительных сколько удалось найти.

### 3.4 Specificity / TNR

$$
Specificity = \frac{TN}{TN+FP}
$$

Доля правильно распознанных отрицательных.

### 3.5 F1-score

$$
F1 = \frac{2\cdot Precision \cdot Recall}{Precision + Recall}
$$

Гармоническое среднее precision и recall.

### 3.6 F$_\beta$-score

$$
F_\beta = (1+\beta^2)\frac{Precision \cdot Recall}{\beta^2 Precision + Recall}
$$

Если $\beta>1$, сильнее важен recall; если $\beta<1$, сильнее важен precision.

### 3.7 Balanced Accuracy

$$
BalancedAccuracy = \frac{TPR + TNR}{2}
$$

Полезна при несбалансированных классах.

### 3.8 MCC — Matthews Correlation Coefficient

$$
MCC=\frac{TP\cdot TN - FP\cdot FN}{\sqrt{(TP+FP)(TP+FN)(TN+FP)(TN+FN)}}
$$

Сильная метрика для бинарной классификации, хорошо работает при дисбалансе.

### 3.9 Cohen’s Kappa

$$
\kappa = \frac{p_o - p_e}{1-p_e}
$$

Учитывает случайное согласие.

### 3.10 LogLoss / Cross-Entropy

Для бинарного случая:

$$
LogLoss = -\frac{1}{n}\sum_{i=1}^{n}\left[y_i\log p_i + (1-y_i)\log(1-p_i)\right]
$$

Оценивает качество вероятностей, а не только класса.

### 3.11 Brier score

$$
Brier = \frac{1}{n}\sum_{i=1}^{n}(p_i-y_i)^2
$$

Оценивает качество вероятностного прогноза и калибровку.

### 3.12 ROC-AUC и PR-AUC

- ROC-AUC показывает, насколько хорошо модель ранжирует положительные выше отрицательных.
- PR-AUC особенно полезна при сильном дисбалансе, когда положительный класс редкий.

## 4. Многоклассовая классификация

Часто используют:
- accuracy;
- macro / micro / weighted precision, recall, F1;
- multiclass log loss;
- top-k accuracy;
- one-vs-rest ROC-AUC.

Macro-average усредняет метрику по классам одинаково.
Micro-average агрегирует TP/FP/FN по всем классам.
Weighted-average учитывает размер классов.

## 5. Ранжирование и поиск

Когда важен не только ответ, но и порядок выдачи.

### 5.1 Precision@k

$$
P@k = \frac{\text{релевантных объектов в топ-}k}{k}
$$

### 5.2 Recall@k

$$
R@k = \frac{\text{релевантных объектов в топ-}k}{\text{всего релевантных}}
$$

### 5.3 MRR — Mean Reciprocal Rank

$$
MRR = \frac{1}{n}\sum_{i=1}^{n}\frac{1}{rank_i}
$$

Где $rank_i$ — позиция первого релевантного объекта.

### 5.4 AP и MAP

$$
MAP = \frac{1}{Q}\sum_{q=1}^{Q} AP(q)
$$

AP — средняя precision по всем позициям релевантных объектов. MAP — среднее по запросам.

### 5.5 DCG и NDCG

$$
DCG@k = \sum_{i=1}^{k}\frac{2^{rel_i}-1}{\log_2(i+1)}
$$

$$
NDCG@k = \frac{DCG@k}{IDCG@k}
$$

Одна из главных метрик ранжирования.

## 6. Кластеризация

Если истинных меток нет, используют внутренние и внешние метрики.

### 6.1 Silhouette score

$$
s = \frac{b-a}{\max(a,b)}
$$

Где $a$ — средняя дистанция до объектов своего кластера, а $b$ — минимальная средняя дистанция до другого кластера.

### 6.2 Davies–Bouldin Index

Чем меньше, тем лучше.

### 6.3 Calinski–Harabasz Index

Чем больше, тем лучше.

### 6.4 Внешние метрики кластеризации

- Purity;
- Rand Index;
- Adjusted Rand Index (ARI);
- NMI.

Используются, если есть истинные классы для сравнения.

## 7. Сегментация изображений

### IoU / Jaccard

$$
IoU = \frac{|A \cap B|}{|A \cup B|}
$$

### Dice coefficient

$$
Dice = \frac{2|A \cap B|}{|A| + |B|}
$$

## 8. Обнаружение объектов

Главные метрики:
- IoU;
- mAP.

mAP — основная метрика object detection.

## 9. Временные ряды

Часто используют MAE, RMSE, MAPE, sMAPE, а также:

$$
MASE = \frac{\text{MAE модели}}{\text{MAE наивного прогноза}}
$$

Если $MASE < 1$, модель лучше наивного прогноза.

## 10. Вероятностный прогноз и калибровка

### NLL
Негативный логарифм правдоподобия.

### CRPS
Оценивает качество предсказанного распределения.

### ECE
Expected Calibration Error — среднее отклонение предсказанной вероятности от фактической частоты.

## 11. Важные выводы для экзамена

- Loss — то, что оптимизируют при обучении.
- Metric — то, чем оценивают качество модели.
- Для регрессии используют MAE, MSE, RMSE, R², MAPE и другие.
- Для классификации используют Accuracy, Precision, Recall, F1, ROC-AUC, PR-AUC, LogLoss.
- Для ранжирования — NDCG, MAP, MRR, Precision@k.
- Для кластеризации — Silhouette, Davies–Bouldin, Calinski–Harabasz, ARI, NMI.

## 12. Разбор экзаменационного вопроса

Задача: предсказать число просмотров статьи.

Это регрессия, потому что целевая переменная — числовая и используется в исходном виде.

Из вариантов:
- F1-score — нет;
- Accuracy — нет;
- MSE — да;
- не подходит ни одна — нет;
- ROC-AUC — нет.

**Правильный ответ: MSE.**
